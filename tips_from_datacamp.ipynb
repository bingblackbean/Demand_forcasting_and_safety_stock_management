* partial and rolling agg

```
import pandas as pd
from random import random
df = pd.DataFrame([random() for i in range(50)])
# use partial to calculate percentiles
from functools import partial
import numpy as np
percentiles = [5,10,25,50,75,90,95]
percentiles_func_list = [partial(np.percentile,q = p) for p in percentiles]
new_df = df.rolling(20).agg(percentiles_func_list)
new_df
```

```
# time shift and create new df
lags_data = {f'lag_{i}':df[0].shift(i) for i in [1,2,3,7,14] }
lags_df = pd.DataFrame(lags_data)
lags_df
```
